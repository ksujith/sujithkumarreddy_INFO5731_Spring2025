```python
import string
import re

def sentences():
    sent_a = input("Enter the first sentence: ")
    sent_b = input("Enter the second sentence: ")
    return sent_a, sent_b

def transform_to_list(sentence):
    "sentenence to words"
    return re.findall('\w+|[.,;:!?]',sentence)

def compare_sentences(list1, list2):
    common_words = set(list1) & set(list2)
    return list(common_words)

def remove_punctuation(sentence):
    return sentence.translate(str.maketrans("", "", string.punctuation))

def count_vowels(sentence):
    vowels = "aeiou"
    count = {vowel: sentence.lower().count(vowel) for vowel in vowels}
    return count

def print_vowel_counts(sentence, count):
    print(f"{sentence}: {', '.join([f'{vowel}:{count[vowel]}' for vowel in count])}")

# (a) Prompt the user to input two sentences
sentence1, sentence2 = sentences()

# (b) Transform each sentence into a list of words
list1 = transform_to_list(sentence1)
list2 = transform_to_list(sentence2)
print(list1+list2)

# (c) Compare the two sentences
common_words = compare_sentences(list1, list2)
print("Common Words:", common_words)

# (d) Remove punctuation from the sentences
sent_chg_a = remove_punctuation(sentence1)
sent_chg_b = remove_punctuation(sentence2)
print("Sanitized Sentence 1:", sent_chg_a)
print("Sanitized Sentence 2:", sent_chg_b)

# (e) Count the number of vowels in each sentence
vowel_count_1 = count_vowels(sent_chg_a)
vowel_count_2 = count_vowels(sent_chg_b)

print_vowel_counts("Sentence 1", vowel_count_1)
print_vowel_counts("Sentence 2", vowel_count_2)
```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
# Your code here
import math

def evaluate_expression(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero"
    except SyntaxError:
        return "Error: Invalid input expression"
    except Exception as e:
        return f"Error: {str(e)}"

def main():
    print("Simple Command-Line Calculator\nTo quit type 'exit'")

    while True:
        user_input = input("Enter a mathematical expression: ")

        # Implement a command to exit the calculator
        if user_input.lower() == 'exit':
            print("Exiting the calculator. Thank You!")
            break

        # Handle parentheses in mathematical expressions
        while '(' in user_input:
            opening_parenthesis = user_input.rfind('(')
            closing_parenthesis = user_input.find(')', opening_parenthesis)
            if opening_parenthesis != -1 and closing_parenthesis != -1:
                sub_expression = user_input[opening_parenthesis + 1:closing_parenthesis]
                result = evaluate_expression(sub_expression)
                user_input = user_input[:opening_parenthesis] + str(result) + user_input[closing_parenthesis + 1:]

        # Additional mathematical operations
        user_input = user_input.replace("^", "**")  # Exponentiation
        user_input = user_input.replace("sqrt", "math.sqrt")  # Square root
        user_input = user_input.replace("%", "/100")  # Modulus

        # Evaluate the mathematical expression
        result = evaluate_expression(user_input)
        print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
import csv
from collections import OrderedDict
from scipy.stats import hmean

def read_csv(file_path):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        return list(reader)

def calculate_differences(data_dict):
    differences = {}
    for year in range(2009, 2015):
        if 'Occupied Housing Units' in data_dict[year] and 'Occupied Housing Units' in data_dict[year - 1]:
            difference = data_dict[year]['Occupied Housing Units'] - data_dict[year - 1]['Occupied Housing Units']
            differences[f"{year-1}-{year}"] = difference
    return differences

def calculate_percentages(data_dict):
    print("\n(b) Percentage of Housing Units: Vacant and Occupied")
    print("Year\tVacant Housing Units\tOccupied Housing Units")
    for year in range(2008, 2015):
        total_units = data_dict[year]['Total Housing Units']
        vacant_units = data_dict[year]['Vacant Housing Units']
        occupied_units = data_dict[year]['Occupied Housing Units']

        vacant_percentage = (vacant_units / total_units) * 100
        occupied_percentage = (occupied_units / total_units) * 100

        print(f"{year}\t{vacant_percentage:.2f}%\t\t\t{occupied_percentage:.2f}%")

def find_highest_values(data_dict):
    print("\n(c) Highest Number of Housing Units:")
    print("Year\t\tValue")
    highest_vacant_year = max(data_dict.keys(), key=lambda year: data_dict[year]['Vacant Housing Units'])
    highest_occupied_year = max(data_dict.keys(), key=lambda year: data_dict[year]['Occupied Housing Units'])
    print(f"Vacant Housing Units\t{highest_vacant_year}\t{data_dict[highest_vacant_year]['Vacant Housing Units']}")
    print(f"Occupied Housing Units\t{highest_occupied_year}\t{data_dict[highest_occupied_year]['Occupied Housing Units']}")

def calculate_harmonic_mean(dataset):
    harmonic_mean = len(dataset) / sum(1/x for x in dataset)
    return round(harmonic_mean, 4)

# File path to 'Assignment1_denton_housing.csv'
file_path = '/content/Assignment1_denton_housing.csv'

# Read the CSV file
data = read_csv(file_path)

# Parse the data into a dictionary
data_dict = OrderedDict()
for row in data:
    year = int(row['year'])
    title_field = row['title_field']
    value = int(row['value']) if row['value'].isdigit() else row['value']

    if year not in data_dict:
        data_dict[year] = {}

    data_dict[year][title_field] = value

# (a)
differences = calculate_differences(data_dict)
print("\n(a) Differences in Occupied Housing Units from Year to Year:")
sorted_differences = sorted(differences.items(), key=lambda x: x[1])
for item in sorted_differences:
    print(f"{item[0]}: {item[1]}")

# (b)
calculate_percentages(data_dict)

# (c)
find_highest_values(data_dict)

# (d)
print("\n(d) Harmonic Mean:")
dataset_for_harmonic_mean = [data_dict[year]['Total Housing Units'] for year in range(2008, 2015)]
harmonic_mean_result = calculate_harmonic_mean(dataset_for_harmonic_mean)
print("Harmonic Mean:", harmonic_mean_result)
```

Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

Original Prompt

```python
import csv
import os

class InventoryManager:
    """
    A class to manage inventory records using a CSV file.

    Features:
    - Adding new products
    - Deleting existing products
    - Updating product information
    - Searching for products by ID
    - Automated restocking if quantity < restock threshold
    - Generating a report (total inventory value + category breakdown)
    - Handling corrupted or missing CSV files
    """

    def __init__(self, filename="inventory.csv"):
        """
        Constructor that sets the CSV filename, expected fields,
        and ensures file integrity at startup.
        """
        self.filename = filename
        self.fields = ["ProductID", "Name", "Category", "Quantity", "Price", "RestockThreshold"]
        self.ensure_file_integrity()

    def ensure_file_integrity(self):
        """
        Verify that the CSV file exists, is not empty, and has correct headers.
        If any issue is found, re-create the file with the default headers.
        """
        # Check if the file doesn't exist or is empty
        if not os.path.exists(self.filename) or os.stat(self.filename).st_size == 0:
            print("Warning: Inventory file is missing or empty. Creating a new file.")
            self.create_csv()
        else:
            try:
                with open(self.filename, mode='r', newline='') as file:
                    reader = csv.reader(file)
                    header = next(reader, None)
                    # If the header is missing or doesn't match expected fields, recreate it
                    if header != self.fields:
                        print("Warning: Corrupted CSV detected. Re-initializing file.")
                        self.create_csv()
            except Exception as e:
                print(f"Error reading inventory file: {e}. Re-initializing file.")
                self.create_csv()

    def create_csv(self):
        """
        Create a new CSV file and write the header row.
        This is called if the file is missing or corrupted.
        """
        try:
            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(self.fields)
        except Exception as e:
            print(f"Error creating new CSV file: {e}")

    def product_exists(self, product_id):
        """
        Check if a product with the given ProductID already exists in the inventory CSV.
        Returns True if it exists, otherwise False.
        """
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)  # Skip header
                for row in reader:
                    if row and row[0] == product_id:
                        return True
        except Exception as e:
            print(f"Error checking product existence: {e}")
        return False

    def add_product(self, product_id, name, category, quantity, price, restock_threshold):
        """
        Add a new product to the inventory CSV if it does not already exist
        and the quantity is valid.
        """
        # Validate for duplicates
        if self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} already exists.")
            return

        # Validate quantity is positive
        if int(quantity) <= 0:
            print(f"Error: Quantity must be greater than zero for ProductID {product_id}.")
            return

        try:
            with open(self.filename, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([product_id, name, category, quantity, price, restock_threshold])
            # After adding, check if we need to restock automatically
            self.auto_restock(product_id)
            print(f"Product {product_id} added successfully.")
        except Exception as e:
            print(f"Error adding product: {e}")

    def delete_product(self, product_id):
        """
        Delete the product with the specified ProductID from the CSV.
        """
        if not self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} does not exist.")
            return

        rows = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                header = next(reader, None)
                rows.append(header)  # Keep the header for rewriting
                for row in reader:
                    # Skip the row if it matches the product to delete
                    if row and row[0] != product_id:
                        rows.append(row)

            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            print(f"Product {product_id} deleted successfully.")
        except Exception as e:
            print(f"Error deleting product: {e}")

    def update_product(self, product_id, field, new_value):
        """
        Update a specific field of an existing product.
        For instance, update_product("101", "Quantity", "15").
        """
        # Validate product existence
        if not self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} does not exist.")
            return

        # Validate field name
        if field not in self.fields:
            print(f"Error: '{field}' is not a valid field.")
            return

        rows = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                header = next(reader, None)
                rows.append(header)
                for row in reader:
                    if row and row[0] == product_id:
                        # Convert to int if updating Quantity/RestockThreshold
                        # Convert to float if updating Price
                        if field in ["Quantity", "RestockThreshold"]:
                            if int(new_value) <= 0:
                                print(f"Error: '{field}' must be > 0.")
                                return
                        row[header.index(field)] = new_value
                    rows.append(row)

            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)

            # Check if updated field triggers an auto restock
            if field in ["Quantity", "RestockThreshold"]:
                self.auto_restock(product_id)

            print(f"Product {product_id} updated successfully.")
        except Exception as e:
            print(f"Error updating product: {e}")

    def search_product(self, product_id):
        """
        Search for a product by ProductID in the CSV and return its record if found.
        """
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)
                for row in reader:
                    if row and row[0] == product_id:
                        return row
        except Exception as e:
            print(f"Error searching for product: {e}")
        return None

    def auto_restock(self, product_id):
        """
        Automatically restock a product if Quantity < RestockThreshold.
        This function reads the product record, checks the threshold,
        and updates Quantity to (RestockThreshold + 10) if necessary.
        """
        record = self.search_product(product_id)
        if record:
            quantity = int(record[3])
            restock_threshold = int(record[5])
            if quantity < restock_threshold:
                # Increase quantity by 10 beyond threshold
                new_quantity = restock_threshold + 10
                self.update_product(product_id, "Quantity", str(new_quantity))
                print(f"Auto-restocked Product {product_id} to Quantity {new_quantity}.")

    def generate_report(self):
        """
        Generate a report showing:
        1. Total Inventory Value
        2. Category-wise breakdown of inventory value
        """
        category_totals = {}
        total_value = 0
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)  # Skip header
                for row in reader:
                    if row:
                        category = row[2]
                        quantity = int(row[3])
                        price = float(row[4])
                        item_value = quantity * price
                        total_value += item_value
                        category_totals[category] = category_totals.get(category, 0) + item_value

            # Display the totals
            print(f"\n--- INVENTORY REPORT ---")
            print(f"Total Inventory Value: ${total_value:,.2f}")
            print("Category Breakdown:")
            for cat, value in category_totals.items():
                print(f"  {cat}: ${value:,.2f}")
            print("------------------------\n")
        except Exception as e:
            print(f"Error generating report: {e}")

    def init_sample_inventory(self):
        """
        Initialize the inventory CSV with 50 diverse products across different categories
        for testing and demonstration purposes.
        """
        # List of 50 sample products with varying categories
        sample_products = [
            # Format: (ProductID, Name, Category, Quantity, Price, RestockThreshold)
            ("A001", "Laptop A", "Electronics", 15, 899.99, 5),
            ("A002", "Laptop B", "Electronics", 10, 1299.99, 5),
            ("A003", "Laptop C", "Electronics", 8, 1599.99, 5),
            ("M101", "Mouse A", "Accessories", 40, 12.99, 10),
            ("M102", "Mouse B", "Accessories", 25, 25.99, 10),
            ("K201", "Keyboard A", "Accessories", 35, 49.99, 10),
            ("K202", "Keyboard B", "Accessories", 20, 69.99, 10),
            ("H301", "Headphones A", "Audio", 18, 79.99, 5),
            ("H302", "Headphones B", "Audio", 12, 99.99, 5),
            ("H303", "Headphones C", "Audio", 10, 129.99, 5),
            ("S401", "Smartphone A", "Electronics", 22, 699.99, 10),
            ("S402", "Smartphone B", "Electronics", 15, 799.99, 10),
            ("S403", "Smartphone C", "Electronics", 18, 999.99, 10),
            ("T501", "Tablet A", "Electronics", 25, 499.99, 5),
            ("T502", "Tablet B", "Electronics", 20, 599.99, 5),
            ("T503", "Tablet C", "Electronics", 15, 699.99, 5),
            ("C601", "Charger A", "Accessories", 30, 14.99, 10),
            ("C602", "Charger B", "Accessories", 20, 19.99, 10),
            ("U701", "USB Drive A", "Storage", 100, 9.99, 20),
            ("U702", "USB Drive B", "Storage", 80, 14.99, 20),
            ("U703", "USB Drive C", "Storage", 60, 19.99, 20),
            ("E801", "External HDD A", "Storage", 15, 89.99, 5),
            ("E802", "External HDD B", "Storage", 12, 119.99, 5),
            ("R901", "Router A", "Networking", 10, 49.99, 5),
            ("R902", "Router B", "Networking", 8, 79.99, 5),
            ("SC01", "Scanner A", "Office", 5, 129.99, 3),
            ("PR01", "Printer A", "Office", 6, 199.99, 3),
            ("PR02", "Printer B", "Office", 4, 299.99, 3),
            ("SP01", "Speaker A", "Audio", 9, 59.99, 5),
            ("SP02", "Speaker B", "Audio", 7, 89.99, 5),
            ("SP03", "Speaker C", "Audio", 5, 129.99, 5),
            ("C603", "Charger C", "Accessories", 25, 29.99, 10),
            ("M103", "Mouse C", "Accessories", 50, 19.99, 10),
            ("K203", "Keyboard C", "Accessories", 14, 39.99, 10),
            ("SD01", "SSD Drive A", "Storage", 20, 59.99, 5),
            ("SD02", "SSD Drive B", "Storage", 10, 99.99, 5),
            ("SD03", "SSD Drive C", "Storage", 12, 149.99, 5),
            ("P101", "Paper A4", "Office", 100, 0.15, 50),
            ("P102", "Paper Letter", "Office", 200, 0.10, 50),
            ("T101", "Toner A", "Office", 7, 49.99, 3),
            ("T102", "Toner B", "Office", 5, 59.99, 3),
            ("E803", "External HDD C", "Storage", 8, 139.99, 5),
            ("B101", "Binder Clips", "Office", 50, 2.99, 10),
            ("B102", "Ballpoint Pen", "Office", 100, 0.25, 20),
            ("N101", "Notebook A", "Office", 30, 1.99, 10),
            ("N102", "Notebook B", "Office", 25, 2.99, 10),
            ("F101", "Flashlight", "Accessories", 10, 24.99, 5),
            ("W101", "Wireless Adapter", "Networking", 12, 29.99, 5),
            ("H304", "Headphones D", "Audio", 8, 109.99, 5),
            ("P103", "Projector", "Office", 3, 399.99, 2)
        ]

        # Recreate the CSV to ensure a clean slate
        self.create_csv()

        # Add each sample product
        for prod in sample_products:
            self.add_product(*prod)

    def run_demo(self):
        """
        Demonstrate a few operations for quick testing.
        This method is optional; you can comment it out or remove it as needed.
        """
        # Add a product that already exists
        self.add_product("A001", "Laptop A", "Electronics", 1, 1000.0, 5)

        # Update a product's price
        self.update_product("A002", "Price", "1499.99")

        # Search an existing product
        result = self.search_product("A002")
        print("Search for Product A002:", result)

        # Delete a product
        self.delete_product("P103")  # Projector

        # Generate a final report
        self.generate_report()


# --------------- Example Usage ---------------

if __name__ == "__main__":
    # Initialize an InventoryManager object
    inventory_manager = InventoryManager("my_inventory.csv")

    # Populate with 50 default products to meet the requirement
    inventory_manager.init_sample_inventory()

    # Optionally run a quick demo of operations
    inventory_manager.run_demo()
```

```python
import csv
import os

class InventoryManager:
    """
    A class to manage inventory records using a CSV file.

    Features:
    - Adding new products
    - Deleting existing products
    - Updating product information
    - Searching for products by ID
    - Automated restocking if quantity < restock threshold
    - Generating a report (total inventory value + category breakdown)
    - Handling corrupted or missing CSV files
    """

    def __init__(self, filename="inventory.csv"):
        """
        Constructor that sets the CSV filename, expected fields,
        and ensures file integrity at startup.
        """
        self.filename = filename
        self.fields = ["ProductID", "Name", "Category", "Quantity", "Price", "RestockThreshold"]
        self.ensure_file_integrity()

    def ensure_file_integrity(self):
        """
        Verify that the CSV file exists, is not empty, and has correct headers.
        If any issue is found, re-create the file with the default headers.
        """
        # Check if the file doesn't exist or is empty
        if not os.path.exists(self.filename) or os.stat(self.filename).st_size == 0:
            print("Warning: Inventory file is missing or empty. Creating a new file.")
            self.create_csv()
        else:
            try:
                with open(self.filename, mode='r', newline='') as file:
                    reader = csv.reader(file)
                    header = next(reader, None)
                    # If the header is missing or doesn't match expected fields, recreate it
                    if header != self.fields:
                        print("Warning: Corrupted CSV detected. Re-initializing file.")
                        self.create_csv()
            except Exception as e:
                print(f"Error reading inventory file: {e}. Re-initializing file.")
                self.create_csv()

    def create_csv(self):
        """
        Create a new CSV file and write the header row.
        This is called if the file is missing or corrupted.
        """
        try:
            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(self.fields)
        except Exception as e:
            print(f"Error creating new CSV file: {e}")

    def product_exists(self, product_id):
        """
        Check if a product with the given ProductID already exists in the inventory CSV.
        Returns True if it exists, otherwise False.
        """
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)  # Skip header
                for row in reader:
                    if row and row[0] == product_id:
                        return True
        except Exception as e:
            print(f"Error checking product existence: {e}")
        return False

    def add_product(self, product_id, name, category, quantity, price, restock_threshold):
        """
        Add a new product to the inventory CSV if it does not already exist
        and the quantity is valid.
        """
        # Validate for duplicates
        if self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} already exists.")
            return

        # Validate quantity is positive
        if int(quantity) <= 0:
            print(f"Error: Quantity must be greater than zero for ProductID {product_id}.")
            return

        try:
            with open(self.filename, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([product_id, name, category, quantity, price, restock_threshold])
            # After adding, check if we need to restock automatically
            self.auto_restock(product_id)
            print(f"Product {product_id} added successfully.")
        except Exception as e:
            print(f"Error adding product: {e}")

    def delete_product(self, product_id):
        """
        Delete the product with the specified ProductID from the CSV.
        """
        if not self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} does not exist.")
            return

        rows = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                header = next(reader, None)
                rows.append(header)  # Keep the header for rewriting
                for row in reader:
                    # Skip the row if it matches the product to delete
                    if row and row[0] != product_id:
                        rows.append(row)

            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            print(f"Product {product_id} deleted successfully.")
        except Exception as e:
            print(f"Error deleting product: {e}")

    def update_product(self, product_id, field, new_value):
        """
        Update a specific field of an existing product.
        For instance, update_product("101", "Quantity", "15").
        """
        # Validate product existence
        if not self.product_exists(product_id):
            print(f"Error: Product with ID {product_id} does not exist.")
            return

        # Validate field name
        if field not in self.fields:
            print(f"Error: '{field}' is not a valid field.")
            return

        rows = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                header = next(reader, None)
                rows.append(header)
                for row in reader:
                    if row and row[0] == product_id:
                        # Convert to int if updating Quantity/RestockThreshold
                        if field in ["Quantity", "RestockThreshold"]:
                            if int(new_value) <= 0:
                                print(f"Error: '{field}' must be > 0.")
                                return
                        row[header.index(field)] = new_value
                    rows.append(row)

            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)

            # Check if updated field triggers an auto restock
            if field in ["Quantity", "RestockThreshold"]:
                self.auto_restock(product_id)

            print(f"Product {product_id} updated successfully.")
        except Exception as e:
            print(f"Error updating product: {e}")

    def search_product(self, product_id):
        """
        Search for a product by ProductID in the CSV and return its record if found.
        """
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)
                for row in reader:
                    if row and row[0] == product_id:
                        return row
        except Exception as e:
            print(f"Error searching for product: {e}")
        return None

    def auto_restock(self, product_id):
        """
        Automatically restock a product if Quantity < RestockThreshold.
        This function reads the product record, checks the threshold,
        and updates Quantity to (RestockThreshold + 10) if necessary.
        """
        record = self.search_product(product_id)
        if record:
            quantity = int(record[3])
            restock_threshold = int(record[5])
            if quantity < restock_threshold:
                # Increase quantity to 10 above threshold
                new_quantity = restock_threshold + 10
                self.update_product(product_id, "Quantity", str(new_quantity))
                print(f"Auto-restocked Product {product_id} to Quantity {new_quantity}.")

    def generate_report(self):
        """
        Generate a report showing:
        1. Total Inventory Value
        2. Category-wise breakdown of inventory value
        """
        category_totals = {}
        total_value = 0
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader, None)  # Skip header
                for row in reader:
                    if row:
                        category = row[2]
                        quantity = int(row[3])
                        price = float(row[4])
                        item_value = quantity * price
                        total_value += item_value
                        category_totals[category] = category_totals.get(category, 0) + item_value

            # Display the totals
            print(f"\n--- INVENTORY REPORT ---")
            print(f"Total Inventory Value: ${total_value:,.2f}")
            print("Category Breakdown:")
            for cat, value in category_totals.items():
                print(f"  {cat}: ${value:,.2f}")
            print("------------------------\n")
        except Exception as e:
            print(f"Error generating report: {e}")

    def init_sample_inventory(self):
        """
        Initialize the inventory CSV with 50 diverse products across different categories
        for testing and demonstration purposes.
        """
        # List of 50 sample products with varying categories
        sample_products = [
            ("A001", "Laptop A", "Electronics", 15, 899.99, 5),
            ("A002", "Laptop B", "Electronics", 10, 1299.99, 5),
            # ... (other products omitted for brevity) ...
            ("P103", "Projector", "Office", 3, 399.99, 2)
        ]

        # Recreate the CSV to ensure a clean slate
        self.create_csv()

        # Add each sample product
        for prod in sample_products:
            self.add_product(*prod)

# ------------------ Testing Duplicate Product IDs ------------------

def test_duplicate_product():
    # Use a test CSV file so as not to interfere with production data.
    test_file = "test_inventory.csv"

    # Remove test file if it exists from previous runs
    if os.path.exists(test_file):
        os.remove(test_file)

    # Initialize the InventoryManager with the test CSV file.
    inventory_manager = InventoryManager(test_file)

    # First addition: This should succeed.
    inventory_manager.add_product("TEST001", "Test Product", "TestCategory", 10, 100.0, 5)

    # Second addition: Attempt to add the same product ID again.
    inventory_manager.add_product("TEST001", "Test Product Duplicate", "TestCategory", 5, 50.0, 5)

    # The expected output is an error message indicating the product already exists.

if __name__ == "__main__":
    print("=== Running Duplicate Product ID Test ===")
    test_duplicate_product()
```

```python
# Cell 1: flight_booking.py code

import datetime

class Flight:
    """
    Represents a flight with a unique identifier, source/destination,
    departure/arrival times, seating capacity, and current reservations.
    """
    def __init__(self, flight_id, source, destination, departure, arrival, capacity):
        """
        Initialize a Flight instance.

        Args:
            flight_id (str): Unique flight identifier.
            source (str): Departure city.
            destination (str): Arrival city.
            departure (datetime.datetime): Departure time.
            arrival (datetime.datetime): Arrival time.
            capacity (int): Seating capacity.

        Raises:
            ValueError: If flight_id is empty, capacity is not positive,
                        or departure is not before arrival.
        """
        if not flight_id or not isinstance(flight_id, str):
            raise ValueError("Flight ID must be a non-empty string.")
        if capacity <= 0:
            raise ValueError("Capacity must be a positive integer.")
        if departure >= arrival:
            raise ValueError("Departure time must be before arrival time.")

        self.flight_id = flight_id
        self.source = source
        self.destination = destination
        self.departure = departure
        self.arrival = arrival
        self.capacity = capacity
        self.reservations = {}  # Mapping: passenger name -> reservation details

    def available_seats(self):
        """
        Return the number of available seats on this flight.

        Returns:
            int: The number of seats available.
        """
        return self.capacity - len(self.reservations)

    def reserve_seat(self, passenger):
        """
        Reserve a seat for a passenger.

        Args:
            passenger (str): Passenger's name.

        Returns:
            bool: True if reservation is successful.

        Raises:
            ValueError: If the passenger name is invalid.
            Exception: If the flight is fully booked or the passenger has already reserved a seat.
        """
        if not passenger or not isinstance(passenger, str):
            raise ValueError("Passenger name must be a non-empty string.")
        if self.available_seats() <= 0:
            raise Exception("Overbooking error: No available seats on flight {}.".format(self.flight_id))
        if passenger in self.reservations:
            raise Exception("Duplicate booking error: Passenger '{}' already booked on flight {}.".format(passenger, self.flight_id))
        self.reservations[passenger] = {"booking_time": datetime.datetime.now()}
        return True

    def release_seat(self, passenger):
        """
        Release a reserved seat for a passenger.

        Args:
            passenger (str): Passenger's name.

        Returns:
            bool: True if the seat is successfully released.

        Raises:
            Exception: If no reservation exists for the passenger.
        """
        if passenger not in self.reservations:
            raise Exception("Cancellation error: No reservation found for passenger '{}' on flight {}.".format(passenger, self.flight_id))
        del self.reservations[passenger]
        return True

    def performance_report(self):
        """
        Generate a performance report for the flight.

        Returns:
            dict: Report including flight details and occupancy rate.
        """
        occupancy_rate = (len(self.reservations) / self.capacity) * 100
        return {
            "flight_id": self.flight_id,
            "source": self.source,
            "destination": self.destination,
            "departure": self.departure,
            "arrival": self.arrival,
            "capacity": self.capacity,
            "booked_seats": len(self.reservations),
            "occupancy_rate": occupancy_rate,
        }

class BookingSystem:
    """
    Manages multiple flights and provides functionalities for flight operations.
    """
    def __init__(self):
        """Initialize the BookingSystem with an empty dictionary of flights."""
        self.flights = {}

    def add_flight(self, flight):
        """
        Add a new flight to the booking system.

        Args:
            flight (Flight): Flight object to be added.

        Returns:
            bool: True if the flight is added successfully.

        Raises:
            Exception: If the flight_id already exists or the departure is in the past.
        """
        if flight.flight_id in self.flights:
            raise Exception("Duplicate flight error: Flight '{}' already exists.".format(flight.flight_id))
        if flight.departure < datetime.datetime.now():
            raise Exception("Invalid flight error: Departure time is in the past for flight '{}'.".format(flight.flight_id))
        self.flights[flight.flight_id] = flight
        return True

    def find_trips(self, source, destination):
        """
        Find flights matching the given source and destination.

        Args:
            source (str): Departure city.
            destination (str): Arrival city.

        Returns:
            list: A list of matching Flight objects.
        """
        matching_flights = []
        for flight in self.flights.values():
            if flight.source.lower() == source.lower() and flight.destination.lower() == destination.lower():
                matching_flights.append(flight)
        return matching_flights

    def reserve_seat(self, flight_id, passenger):
        """
        Reserve a seat for a passenger on a given flight.

        Args:
            flight_id (str): Flight identifier.
            passenger (str): Passenger's name.

        Returns:
            bool: True if reservation is successful.

        Raises:
            Exception: If the flight does not exist.
        """
        if flight_id not in self.flights:
            raise Exception("Invalid flight error: Flight '{}' does not exist.".format(flight_id))
        return self.flights[flight_id].reserve_seat(passenger)

    def release_seat(self, flight_id, passenger):
        """
        Release a reserved seat for a passenger on a given flight.

        Args:
            flight_id (str): Flight identifier.
            passenger (str): Passenger's name.

        Returns:
            bool: True if the seat is released successfully.

        Raises:
            Exception: If the flight does not exist.
        """
        if flight_id not in self.flights:
            raise Exception("Invalid flight error: Flight '{}' does not exist.".format(flight_id))
        return self.flights[flight_id].release_seat(passenger)

    def generate_report(self, flight_id=None):
        """
        Generate a performance report.

        Args:
            flight_id (str, optional): Specific flight identifier. If provided, returns a report for that flight.
                                        Otherwise, returns reports for all flights.

        Returns:
            dict: Performance report(s).

        Raises:
            Exception: If the specified flight does not exist.
        """
        if flight_id:
            if flight_id not in self.flights:
                raise Exception("Invalid flight error: Flight '{}' does not exist.".format(flight_id))
            return self.flights[flight_id].performance_report()
        else:
            reports = {}
            for fid, flight in self.flights.items():
                reports[fid] = flight.performance_report()
            return reports

    def advanced_search(self, source=None, destination=None, departure_after=None, departure_before=None):
        """
        Perform an advanced search for flights using multiple filters.

        Args:
            source (str, optional): Departure city.
            destination (str, optional): Arrival city.
            departure_after (datetime.datetime, optional): Earliest departure time.
            departure_before (datetime.datetime, optional): Latest departure time.

        Returns:
            list: A list of Flight objects matching the filters.
        """
        results = []
        for flight in self.flights.values():
            if source and flight.source.lower() != source.lower():
                continue
            if destination and flight.destination.lower() != destination.lower():
                continue
            if departure_after and flight.departure < departure_after:
                continue
            if departure_before and flight.departure > departure_before:
                continue
            results.append(flight)
        return results

def parse_datetime(date_str):
    """
    Convert a string in 'YYYY-MM-DD HH:MM' format to a datetime object.

    Args:
        date_str (str): The date/time string.

    Returns:
        datetime.datetime: The corresponding datetime object.

    Raises:
        ValueError: If the date_str format is incorrect.
    """
    try:
        return datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M')
    except Exception as e:
        raise ValueError("Invalid date/time format. Expected 'YYYY-MM-DD HH:MM'. Error: {}".format(e))

def initialize_sample_flights(booking_system):
    """
    Populate the booking system with 50 sample flights.

    Args:
        booking_system (BookingSystem): The booking system to populate.
    """
    base_date = datetime.datetime.now() + datetime.timedelta(days=1)
    for i in range(1, 51):
        flight_id = f"FL{i:03d}"
        source = "CityA" if i % 2 == 0 else "CityB"
        destination = "CityC" if i % 3 == 0 else "CityD"
        departure = base_date + datetime.timedelta(hours=i)
        arrival = departure + datetime.timedelta(hours=2)
        capacity = 10 + (i % 5) * 5
        try:
            flight = Flight(flight_id, source, destination, departure, arrival, capacity)
            booking_system.add_flight(flight)
        except Exception as e:
            print(f"Error adding flight {flight_id}: {e}")

print("Flight booking system code loaded.")
```

# Test Case 15 – Reservation After Cancellation

```python

system = BookingSystem()
initialize_sample_flights(system)
flight_id = "FL007"
try:
    # Ensure the flight exists and reserve a seat.
    if flight_id not in system.flights:
        departure = datetime.datetime.now() + datetime.timedelta(days=2)
        arrival = departure + datetime.timedelta(hours=2)
        flight = Flight(flight_id, "CityG", "CityH", departure, arrival, 15)
        system.add_flight(flight)
    system.reserve_seat(flight_id, "Diana")
    # Simulate cancellation by removing the flight.
    del system.flights[flight_id]
    # Attempt to reserve on the cancelled flight.
    system.reserve_seat(flight_id, "AnotherPassenger")
    print("Test 15 Failed: Reservation succeeded on a cancelled flight.")
except Exception as e:
    print("Test 15 Passed:", e)
```
